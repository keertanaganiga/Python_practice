```
import matplotlib.pyplot as plt
import numpy as np

y=np.array([25,35,35,24,56,67,86])
mylables=['RM','JIN','SUGA','JHOPE','JIMIN','KIM TAEHYUNG','JUNGKOOK']
myexplodes=[0,0,0,0,0,0,0.1]

plt.pie(y,labels=mylables,explode=myexplodes,shadow=True)
plt.show()
```
